## Training Only Scaling on Aurora

```python
class SGSTraining:
    def __init__(self,path,gpu_list):
        self.base_path = path
        self.gpu_list = gpu_list
        self.n_gpu = len(gpu_list)
        self.train_fom = {
            'train_tot': np.zeros((self.n_gpu,4)),
            'train_loop': np.zeros((self.n_gpu,4)), 
            'throughput': np.zeros((self.n_gpu,4)),
            'throughput_par': np.zeros((self.n_gpu,)),
            'train_iter_tot': np.zeros((self.n_gpu,4)),
            'train_iter_avg': np.zeros((self.n_gpu,4)),
        }
        
    def get_value(self,string):
        val = string.split(":")[1].split(",")[-1]
        return float(val.split('=')[-1].strip())
    
    def get_values(self,string):
        vals = string.split(":")[1].split(",")
        arr = np.zeros((4,))
        for i in range(4):
            arr[i] = float(vals[i].split('=')[-1].strip())
        return arr
    
    def parse_files(self):
        counts = np.zeros((self.n_gpu,))
        for i in range(self.n_gpu):
            path = self.base_path+f"/{self.gpu_list[i]}/cfdml_train.o*"
            # Loop over runs found
            run_files = glob.glob(path)
            for run_file in run_files:
                print('Reading file: ', run_file)
                with open(run_file,'r') as fh:
                    for l in fh:
                        if "Total training time" in l:
                            self.train_fom['train_tot'][i] += self.get_values(l)
                            counts[i] += 1
                        if "Total time spent in training loop" in l:
                            self.train_fom['train_loop'][i] += self.get_values(l)
                        if "Total training throughput" in l:
                            self.train_fom['throughput'][i] += self.get_values(l)
                            self.train_fom['throughput_par'][i] += self.get_value(l)
                        if "Total time spent in batch computation" in l:
                            self.train_fom['train_iter_tot'][i] += self.get_values(l)
                        if "Average time for a single batch computation" in l:
                            self.train_fom['train_iter_avg'][i] += self.get_values(l)


        # Divide by the counts for each gpu number to get the average over runs
        for key in self.train_fom.keys():
            val = self.train_fom[key]
            if len(val.shape)==1:
                np.divide(val, counts, out=self.train_fom[key], where=counts>0)
                #self.train_fom[key] = np.divide(val, counts, where=counts>0)
            else:
                for j in range(val.shape[1]):
                    np.divide(val[:,j], counts, out=self.train_fom[key][:,j], where=counts>0)
                    #self.train_fom[key][:,j] = np.divide(val[:,j], counts, where=counts>0)
```

## Full workflow scaling on Aurora

```python
class Workflow:
    def __init__(self,path,node_list,jobID):
        self.base_path = path
        self.node_list = node_list
        self.jobID = jobID
        self.n_nodes = len(node_list)
        self.sim_runtime = np.zeros((self.n_nodes,))
        self.sim_steps = np.zeros((self.n_nodes,))
        self.sim_sr_init = np.zeros((self.n_nodes,))
        self.sim_sr_meta = np.zeros((self.n_nodes,))
        self.sim_sr_train = np.zeros((self.n_nodes,))
        self.sim_sr_train_avg = np.zeros((self.n_nodes,))
        self.sim_ksp_solve = np.zeros((self.n_nodes,))
        self.sim_tsstep = np.zeros((self.n_nodes,))
        self.sim_matmult = np.zeros((self.n_nodes,))
        self.sim_vecmdot = np.zeros((self.n_nodes,))
        self.sim_vecnorm = np.zeros((self.n_nodes,))
        self.sim_ceedop = np.zeros((self.n_nodes,))
        
        self.ml_runtime = np.zeros((self.n_nodes,))
        self.ml_sr_init = np.zeros((self.n_nodes,))
        self.ml_sr_meta = np.zeros((self.n_nodes,))
        self.ml_sr_train = np.zeros((self.n_nodes,))

    def get_sim_runtime(self,string):
        return float(string.split(":")[1].strip())
    
    def get_sim_steps(self,string):
        return float(string.split("step")[1].lstrip().split(' ')[0])
    
    def get_sim_sr_time(self,string,keyword):
        tmp = string.strip(keyword).lstrip()
        return list(map(float,tmp.split()))[2]
    
    def get_sim_kernel_time(self,string,keyword):
        tmp = string.strip(keyword).lstrip()
        tmp = list(map(float,tmp.split()))
        return tmp[2]/tmp[0]
    
    def get_ml_time(self,string):
        tmp = string.split(":")[1].strip()
        tmp = tmp.split(",")[1]
        return float(tmp.split("=")[1].strip())

    def parse_files(self):
        counts = np.zeros((self.n_nodes,))
        for i in range(self.n_nodes):
            path = self.base_path+f"/{self.node_list[i]}/cfdml/{self.jobID}*"
            # Loop over runs found
            run_files = glob.glob(path)
            for file in run_files:
                # Simulation performance
                fpath = file+"/sim/sim.out"
                fh = open(fpath,'r')
                in_fluids_solve_stage = False
                for l in fh:
                    if "Event Stage 2: Fluids Solve" in l:
                        in_fluids_solve_stage = True
                    if "Time taken for solution (sec):" in l:
                        self.sim_runtime[i] += self.get_sim_runtime(l)
                        counts[i] += 1
                    if "Time integrator CONVERGED_" in l:
                        self.sim_steps[i] += self.get_sim_steps(l)
                    if "SmartRedis_Init" in l:
                        self.sim_sr_init[i] += self.get_sim_sr_time(l,"SmartRedis_Init")
                    if "SmartRedis_Meta" in l:
                        self.sim_sr_meta[i] += self.get_sim_sr_time(l,"SmartRedis_Meta")
                    if "SmartRedis_Train" in l:
                        self.sim_sr_train[i] += self.get_sim_sr_time(l,"SmartRedis_Train")
                        self.sim_sr_train_avg[i] += self.get_sim_kernel_time(l,"SmartRedis_Train")
                    if "KSPSolve" in l and in_fluids_solve_stage:
                        self.sim_ksp_solve[i] += self.get_sim_kernel_time(l,"KSPSolve")
                    if "TSStep" in l and in_fluids_solve_stage:
                        self.sim_tsstep[i] += self.get_sim_kernel_time(l,"TSStep")
                    if "MatMult" in l and in_fluids_solve_stage:
                        self.sim_matmult[i] += self.get_sim_kernel_time(l,"MatMult")
                    if "VecMDot" in l and in_fluids_solve_stage:
                        self.sim_vecmdot[i] += self.get_sim_kernel_time(l,"VecMDot")
                    if "VecNorm" in l and in_fluids_solve_stage:
                        self.sim_vecnorm[i] += self.get_sim_kernel_time(l,"VecNorm")
                    if "CeedOpApply" in l and in_fluids_solve_stage:
                        self.sim_ceedop[i] += self.get_sim_kernel_time(l,"CeedOpApply")
    
                # Training performance
                fpath = file+"/train/train.out"
                fh = open(fpath,'r')
                for l in fh:
                    if "Total training time" in l:
                        self.ml_runtime[i] += self.get_ml_time(l)
                    if "SmartRedis client initialization" in l:
                        self.ml_sr_init[i] += self.get_ml_time(l)
                    if "SmartRedis metadata transfer" in l:
                        self.ml_sr_meta[i] += self.get_ml_time(l)
                    if "SmartRedis training batch data transfer" in l:
                        self.ml_sr_train[i] += self.get_ml_time(l)
                        
        # Divide by the counts for each node number to get the average over runs
        np.divide(self.sim_runtime,counts,out=self.sim_runtime,where=counts>0)
        np.divide(self.sim_steps,counts,out=self.sim_steps,where=counts>0)
        np.divide(self.sim_sr_init,counts,out=self.sim_sr_init,where=counts>0)
        np.divide(self.sim_sr_meta,counts,out=self.sim_sr_meta,where=counts>0)
        np.divide(self.sim_sr_train,counts,out=self.sim_sr_train,where=counts>0)
        np.divide(self.sim_sr_train_avg,counts,out=self.sim_sr_train_avg,where=counts>0)
        np.divide(self.sim_ksp_solve,counts,out=self.sim_ksp_solve,where=counts>0)
        np.divide(self.sim_tsstep,counts,out=self.sim_tsstep,where=counts>0)
        np.divide(self.sim_matmult,counts,out=self.sim_matmult,where=counts>0)
        np.divide(self.sim_vecmdot,counts,out=self.sim_vecmdot,where=counts>0)
        np.divide(self.sim_vecnorm,counts,out=self.sim_vecnorm,where=counts>0)
        np.divide(self.sim_ceedop,counts,out=self.sim_ceedop,where=counts>0)
        
        np.divide(self.ml_runtime,counts,out=self.ml_runtime,where=counts>0)
        np.divide(self.ml_sr_init,counts,out=self.ml_sr_init,where=counts>0)
        np.divide(self.ml_sr_meta,counts,out=self.ml_sr_meta,where=counts>0)
        np.divide(self.ml_sr_train,counts,out=self.ml_sr_train,where=counts>0)
        
    def compute_scaling_efficiency(self):
        self.sim_runtime_eta = self.sim_runtime/self.sim_runtime[0]
        self.sim_steps_eta = self.sim_steps/self.sim_steps[0]
        self.sim_sr_init_eta = self.sim_sr_init/self.sim_sr_init[0]
        self.sim_sr_meta_eta = self.sim_sr_meta/self.sim_sr_meta[0]
        self.sim_sr_train_eta = self.sim_sr_train/self.sim_sr_train[0]
        self.sim_sr_train_avg_eta = self.sim_sr_train_avg/self.sim_sr_train_avg[0]
        self.sim_ksp_solve_eta = self.sim_ksp_solve/self.sim_ksp_solve[0]
        self.sim_tsstep_eta = self.sim_tsstep/self.sim_tsstep[0]
        self.sim_matmult_eta = self.sim_matmult/self.sim_matmult[0]
        self.sim_vecmdot_eta = self.sim_vecmdot/self.sim_vecmdot[0]
        self.sim_vecnorm_eta = self.sim_vecnorm/self.sim_vecnorm[0]
        self.sim_ceedop_eta = self.sim_ceedop/self.sim_ceedop[0]
        
        self.ml_runtime_eta = self.ml_runtime/self.ml_runtime[0]
        self.ml_sr_init_eta = self.ml_sr_init/self.ml_sr_init[0]
        self.ml_sr_meta_eta = self.ml_sr_meta/self.ml_sr_meta[0]
        self.ml_sr_train_eta = self.ml_sr_train/self.ml_sr_train[0]
```

## libCEED only Scaling -- Strong Bump Q3 and Q4

```python
class LibCEED:
    def __init__(self,path,node_list):
        self.base_path = path
        self.node_list = node_list
        self.n_nodes = len(node_list)
        self.sim_runtime = np.zeros((self.n_nodes,))
        self.sim_steps = np.zeros((self.n_nodes,))
        self.sim_ksp_solve = np.zeros((self.n_nodes,))
        self.sim_tsstep = np.zeros((self.n_nodes,))
        self.sim_matmult = np.zeros((self.n_nodes,))
        self.sim_vecmdot = np.zeros((self.n_nodes,))
        self.sim_vecnorm = np.zeros((self.n_nodes,))
        self.sim_ceedop = np.zeros((self.n_nodes,))

    def get_sim_runtime(self,string):
        return float(string.split(":")[1].strip())
    
    def get_sim_steps(self,string):
        return float(string.split("step")[1].lstrip().split(' ')[0])
    
    def get_sim_sr_time(self,string,keyword):
        tmp = string.strip(keyword).lstrip()
        return list(map(float,tmp.split()))[2]
    
    def get_sim_kernel_time(self,string,keyword):
        tmp = string.strip(keyword).lstrip()
        tmp = list(map(float,tmp.split()[:8]))
        return tmp[2]/tmp[0]

    def parse_files(self):
        counts = np.zeros((self.n_nodes,))
        for i in range(self.n_nodes):
            path = self.base_path+f"/{self.node_list[i]}/libceed*"
            # Loop over runs found
            run_files = glob.glob(path)
            for file in run_files:
                # Simulation performance
                #fpath = file+"/sim/sim.out"
                fh = open(file,'r')
                in_fluids_solve_stage = False
                for l in fh:
                    if "Event Stage 1: Fluids Solve" in l:
                        in_fluids_solve_stage = True
                    if "Time taken for solution (sec):" in l:
                        self.sim_runtime[i] += self.get_sim_runtime(l)
                        counts[i] += 1
                    if "Time integrator CONVERGED_" in l:
                        self.sim_steps[i] += self.get_sim_steps(l)
                    if "KSPSolve" in l and in_fluids_solve_stage:
                        self.sim_ksp_solve[i] += self.get_sim_kernel_time(l,"KSPSolve")
                    if "TSStep" in l and in_fluids_solve_stage:
                        self.sim_tsstep[i] += self.get_sim_kernel_time(l,"TSStep")
                    if "MatMult" in l and in_fluids_solve_stage:
                        self.sim_matmult[i] += self.get_sim_kernel_time(l,"MatMult")
                    if "VecMDot" in l and in_fluids_solve_stage:
                        self.sim_vecmdot[i] += self.get_sim_kernel_time(l,"VecMDot")
                    if "VecNorm" in l and in_fluids_solve_stage:
                        self.sim_vecnorm[i] += self.get_sim_kernel_time(l,"VecNorm")
                    if "CeedOpApply" in l and in_fluids_solve_stage:
                        self.sim_ceedop[i] += self.get_sim_kernel_time(l,"CeedOpApply")
                        
        # Divide by the counts for each node number to get the average over runs
        np.divide(self.sim_runtime,counts,out=self.sim_runtime,where=counts>0)
        np.divide(self.sim_steps,counts,out=self.sim_steps,where=counts>0)
        np.divide(self.sim_ksp_solve,counts,out=self.sim_ksp_solve,where=counts>0)
        np.divide(self.sim_tsstep,counts,out=self.sim_tsstep,where=counts>0)
        np.divide(self.sim_matmult,counts,out=self.sim_matmult,where=counts>0)
        np.divide(self.sim_vecmdot,counts,out=self.sim_vecmdot,where=counts>0)
        np.divide(self.sim_vecnorm,counts,out=self.sim_vecnorm,where=counts>0)
        np.divide(self.sim_ceedop,counts,out=self.sim_ceedop,where=counts>0)
        
    def compute_scaling_efficiency(self):
        self.sim_runtime_eta = self.sim_runtime/self.sim_runtime[0]
        self.sim_steps_eta = self.sim_steps/self.sim_steps[0]
        self.sim_ksp_solve_eta = self.sim_ksp_solve/self.sim_ksp_solve[0]
        self.sim_tsstep_eta = self.sim_tsstep/self.sim_tsstep[0]
        self.sim_matmult_eta = self.sim_matmult/self.sim_matmult[0]
        self.sim_vecmdot_eta = self.sim_vecmdot/self.sim_vecmdot[0]
        self.sim_vecnorm_eta = self.sim_vecnorm/self.sim_vecnorm[0]
        self.sim_ceedop_eta = self.sim_ceedop/self.sim_ceedop[0]
```

## libCEED only Scaling -- Weak Flat Plate, Q1 and Q3

```python
class LibCEED:
    def __init__(self,path,node_list):
        self.base_path = path
        self.node_list = node_list
        self.n_nodes = len(node_list)
        self.sim_runtime = np.zeros((self.n_nodes,))
        self.sim_steps = np.zeros((self.n_nodes,))
        self.sim_ksp_solve = np.zeros((self.n_nodes,))
        self.sim_tsstep = np.zeros((self.n_nodes,))
        self.sim_matmult = np.zeros((self.n_nodes,))
        self.sim_vecmdot = np.zeros((self.n_nodes,))
        self.sim_vecnorm = np.zeros((self.n_nodes,))
        self.sim_ceedop = np.zeros((self.n_nodes,))

    def get_sim_runtime(self,string):
        return float(string.split(":")[1].strip())
    
    def get_sim_steps(self,string):
        return float(string.split("step")[1].lstrip().split(' ')[0])
    
    def get_sim_sr_time(self,string,keyword):
        tmp = string.strip(keyword).lstrip()
        return list(map(float,tmp.split()))[2]
    
    def get_sim_kernel_time(self,string,keyword):
        tmp = string.strip(keyword).lstrip()
        tmp = list(map(float,tmp.split()[:8]))
        return tmp[2]/tmp[0]

    def parse_files(self):
        counts = np.zeros((self.n_nodes,))
        for i in range(self.n_nodes):
            path = self.base_path+f"/{self.node_list[i]}/libceed*"
            # Loop over runs found
            run_files = glob.glob(path)
            for file in run_files:
                # Simulation performance
                #fpath = file+"/sim/sim.out"
                fh = open(file,'r')
                in_fluids_solve_stage = False
                for l in fh:
                    if "Event Stage 2: Fluids Solve" in l:
                        in_fluids_solve_stage = True
                    if "Time taken for solution (sec):" in l:
                        self.sim_runtime[i] += self.get_sim_runtime(l)
                        counts[i] += 1
                    if "Time integrator CONVERGED_" in l:
                        self.sim_steps[i] += self.get_sim_steps(l)
                    if "KSPSolve" in l and in_fluids_solve_stage:
                        self.sim_ksp_solve[i] += self.get_sim_kernel_time(l,"KSPSolve")
                    if "TSStep" in l and in_fluids_solve_stage:
                        self.sim_tsstep[i] += self.get_sim_kernel_time(l,"TSStep")
                    if "MatMult" in l and in_fluids_solve_stage:
                        self.sim_matmult[i] += self.get_sim_kernel_time(l,"MatMult")
                    if "VecMDot" in l and in_fluids_solve_stage:
                        self.sim_vecmdot[i] += self.get_sim_kernel_time(l,"VecMDot")
                    if "VecNorm" in l and in_fluids_solve_stage:
                        self.sim_vecnorm[i] += self.get_sim_kernel_time(l,"VecNorm")
                    if "CeedOpApply" in l and in_fluids_solve_stage:
                        self.sim_ceedop[i] += self.get_sim_kernel_time(l,"CeedOpApply")
                        
        # Divide by the counts for each node number to get the average over runs
        np.divide(self.sim_runtime,counts,out=self.sim_runtime,where=counts>0)
        np.divide(self.sim_steps,counts,out=self.sim_steps,where=counts>0)
        np.divide(self.sim_ksp_solve,counts,out=self.sim_ksp_solve,where=counts>0)
        np.divide(self.sim_tsstep,counts,out=self.sim_tsstep,where=counts>0)
        np.divide(self.sim_matmult,counts,out=self.sim_matmult,where=counts>0)
        np.divide(self.sim_vecmdot,counts,out=self.sim_vecmdot,where=counts>0)
        np.divide(self.sim_vecnorm,counts,out=self.sim_vecnorm,where=counts>0)
        np.divide(self.sim_ceedop,counts,out=self.sim_ceedop,where=counts>0)
        
    def compute_scaling_efficiency(self):
        self.sim_runtime_eta = self.sim_runtime/self.sim_runtime[0]
        self.sim_steps_eta = self.sim_steps/self.sim_steps[0]
        self.sim_ksp_solve_eta = self.sim_ksp_solve/self.sim_ksp_solve[0]
        self.sim_tsstep_eta = self.sim_tsstep/self.sim_tsstep[0]
        self.sim_matmult_eta = self.sim_matmult/self.sim_matmult[0]
        self.sim_vecmdot_eta = self.sim_vecmdot/self.sim_vecmdot[0]
        self.sim_vecnorm_eta = self.sim_vecnorm/self.sim_vecnorm[0]
        self.sim_ceedop_eta = self.sim_ceedop/self.sim_ceedop[0]
```

```python
# Weak scaling, Flat Plate Q1
top_dir = "/Users/rbalin/Documents/Research/ALCF/CFDML/train_scaling/Aurora/frameworks_2024.2.1/libCEED/weak_scale/sim_only/q1/"
node_list = [128, 256, 512, 1024, 2048]
plate_q1 = LibCEED(top_dir,node_list)
plate_q1.parse_files()
print(plate_q1.sim_runtime, plate_q1.sim_steps)
plate_q1.compute_scaling_efficiency()
print(plate_q1.sim_runtime_eta)
print(plate_q1.sim_tsstep)

# Weak scaling, Flat Plate Q3
top_dir = "/Users/rbalin/Documents/Research/ALCF/CFDML/train_scaling/Aurora/frameworks_2024.2.1/libCEED/weak_scale/sim_only/q3/"
node_list = [128, 256, 512, 1024, 2048]
plate_q3 = LibCEED(top_dir,node_list)
plate_q3.parse_files()
print(plate_q3.sim_runtime, plate_q3.sim_steps)
plate_q3.compute_scaling_efficiency()
print(plate_q3.sim_runtime_eta)
print(plate_q3.sim_tsstep)
```
