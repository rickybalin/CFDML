# Plot the Training Curves of the Models

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
base_path = "/Users/rbalin/Documents/Research/ALCF_PostDoc/Conferences/PASC23/FlatPlate/Train/CRS_6-15_4d/models/"
```

```python
fname = base_path + "3x/NNmodel_log.dat"
base_3x = np.loadtxt(fname)

fname = base_path + "3x/3layers/NNmodel_50n_log.dat"
test_3x = np.loadtxt(fname)

fname = base_path + "6x/NNmodel_log.dat"
base_6x = np.loadtxt(fname)

#fname = base_path + "12x/NNmodel_log.dat"
#base_12x = np.loadtxt(fname)

fname = base_path + "all/NNmodel_log.dat"
base_all = np.loadtxt(fname)
```

```python
epochs = [i for i in range(250)]
fig, axs = plt.subplots(nrows=1, ncols=1, figsize=(5, 4))
axs.plot(epochs, base_3x[:,0], 'd', label="Training")
axs.plot(epochs, base_3x[:,1], 's', label="Validation")
axs.set_yscale("log")
axs.grid()
fig.tight_layout(pad=3.0)
axs.set_ylabel('MSE')
axs.set_xlabel('Epoch')
axs.set_title('Training Curve')
axs.legend()
#plt.savefig(fig_name+"_yerrors.png", dpi='figure', format="png")
plt.show()
```
